```python
%pip install matplotlib seaborn

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {    'Gender': ['Male', 'Female', 'Male', 'Male', 'Female', 'Female'],
    'Education Level': ['High School', 'College', 'High School', 'College', 'Graduate', 'High School'] }
```

```python
df = pd.DataFrame(data)
```

```python
crosstab_table = pd.crosstab(df['Gender'], df['Education Level']) 
```

```python
plt.figure(figsize=(8, 6))
crosstab_table.plot(kind='bar', stacked=True)
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Crosstab: Gender vs Education Level')
plt.legend(title='Education Level')
plt.show() 
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(crosstab_table, annot=True, cmap='YlGnBu', fmt='d')
plt.xlabel('Education Level')
plt.ylabel('Gender')
plt.title('Crosstab Heatmap: Gender vs Education Level')
plt.show()
```
