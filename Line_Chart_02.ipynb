```python
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

# Sample medical transcripts categorized into different types of reviews
reviews = {
    "Symptoms": [
        "Patient presented with symptoms of cough and shortness of breath.",
        "Patient complains of chest pain and difficulty breathing."
    ],
    "Diagnosis": [
        "Diagnosis confirmed as bronchitis, prescribed antibiotics and inhaler.",
        "X-ray shows signs of pneumonia in the left lung."
    ],
    "Treatment": [
        "Treatment plan includes rest, hydration, and pain management.",
        "Administered IV fluids and antibiotics."
    ],
    "Follow-up": [
        "Patient discharged with instructions to follow up in one week.",
        "Scheduled for follow-up appointment with primary care physician."
    ],
    "Health History": [
        "Patient history includes hypertension and diabetes.",
        "Family history of heart disease and asthma."
    ],
    "Procedures": [
        "Performed ECG to monitor heart function.",
        "Patient referred to cardiologist for further evaluation."
    ]
}

# Calculate word frequencies for each category of reviews
word_frequencies = {}
for category, transcript_list in reviews.items():
    all_words = []
    for transcript in transcript_list:
        tokens = word_tokenize(transcript.lower())
        all_words.extend(tokens)
    fdist = FreqDist(all_words)
    word_frequencies[category] = fdist

# Create line chart
plt.figure(figsize=(10, 6))
for category, fdist in word_frequencies.items():
    sorted_words = sorted(fdist.items(), key=lambda x: x[1], reverse=True)[:10]  # Top 10 frequent words
    words, frequencies = zip(*sorted_words)
    plt.plot(words, frequencies, marker='o', label=category)

plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Word Frequencies Across Different Categories of Medical Reviews')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
