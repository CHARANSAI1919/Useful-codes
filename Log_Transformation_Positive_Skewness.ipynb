```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = np.random.gamma(2, 1.5, 1000)  


df = pd.DataFrame(data, columns=['Skewed_Data'])


plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title('Before Transformation')
plt.hist(df['Skewed_Data'], bins=30, color='skyblue')
plt.xlabel('Values')
plt.ylabel('Frequency')


df['Log_Transformed'] = np.log1p(df['Skewed_Data'])  # Log transformation


plt.subplot(1, 2, 2)
plt.title('After Log Transformation')
plt.hist(df['Log_Transformed'], bins=30, color='orange')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
