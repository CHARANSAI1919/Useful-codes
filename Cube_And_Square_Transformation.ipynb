```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = np.random.chisquare(2, 1000)  


df = pd.DataFrame(data, columns=['Skewed_Data'])


plt.figure(figsize=(10, 4))

plt.subplot(1, 4, 1)
plt.title('Before Transformation')
plt.hist(df['Skewed_Data'], bins=30, color='skyblue')
plt.xlabel('Values')
plt.ylabel('Frequency')


df['Square_Root_Transformed'] = np.sqrt(df['Skewed_Data'])

# Visualize the square root transformed data
plt.subplot(1, 4, 2)
plt.title('After Square Root Transformation')
plt.hist(df['Square_Root_Transformed'], bins=30, color='orange')
plt.xlabel('Values')
plt.ylabel('Frequency')

# Applying cube root transformation to mitigate the negatively skewed data
df['Cube_Root_Transformed'] = np.cbrt(df['Skewed_Data'])

# Visualize the cube root transformed data
plt.subplot(1, 4, 3)
plt.title('After Cube Root Transformation')
plt.hist(df['Cube_Root_Transformed'], bins=30, color='green')
plt.xlabel('Values')
plt.ylabel('Frequency')

# Applying reciprocal transformation to mitigate the negatively skewed data
df['Reciprocal_Transformed'] = 1 / df['Skewed_Data']

# Visualize the reciprocal transformed data
plt.subplot(1, 4, 4)
plt.title('After Reciprocal Transformation')
plt.hist(df['Reciprocal_Transformed'], bins=30, color='red')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating a dataset using the Exponential Distribution
data = np.random.exponential(scale=0.5, size=1000)  # Adjust scale for skewness

# Visualize the negatively skewed data
plt.figure(figsize=(6, 4))
plt.hist(data, bins=30, color='skyblue')
plt.title('Negatively Skewed Data (Exponential Distribution)')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.exponential(scale=1, size=5000)

transformed_data = -5 * (data / data.max())

final_data = transformed_data[(transformed_data >= -5) & (transformed_data <= 0)][:1000]


df = pd.DataFrame(final_data, columns=['Skewed_Data'])

plt.figure(figsize=(10, 4))

plt.subplot(1, 4, 1)
plt.title('Before Transformation')
plt.hist(df['Skewed_Data'], bins=30, color='skyblue')
plt.xlabel('Values')
plt.ylabel('Frequency')


df['Square_Root_Transformed'] = np.sqrt(df['Skewed_Data'])


plt.subplot(1, 4, 2)
plt.title('After Square Root Transformation')
plt.hist(df['Square_Root_Transformed'], bins=30, color='orange')
plt.xlabel('Values')
plt.ylabel('Frequency')


df['Cube_Root_Transformed'] = np.cbrt(df['Skewed_Data'])


plt.subplot(1, 4, 3)
plt.title('After Cube Root Transformation')
plt.hist(df['Cube_Root_Transformed'], bins=30, color='green')
plt.xlabel('Values')
plt.ylabel('Frequency')


df['Reciprocal_Transformed'] = 1 / df['Skewed_Data']


plt.subplot(1, 4, 4)
plt.title('After Reciprocal Transformation')
plt.hist(df['Reciprocal_Transformed'], bins=30, color='red')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
