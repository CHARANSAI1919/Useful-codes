```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

# Example data
data = {
    "feature1": [10, 20, 30, 40, 50],
    "feature2": [5, 15, 25, 35, 45]
}

# Create the DataFrame
data_frame = pd.DataFrame(data)

# Display the DataFrame
print(data_frame)
```

```python
covariance_matrix = np.cov(data_frame['feature1'], data_frame['feature2'])
covariance = covariance_matrix[0, 1]
print(covariance)
```

```python
correlation_matrix = data_frame[['feature1', 'feature2']].corr()
correlation = correlation_matrix.loc['feature1', 'feature2']
print(correlation)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data_frame['feature1'], data_frame['feature2'])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Feature 1 vs Feature 2')
plt.show() 
```
